```python
import pandas as pd
```

```python
heart_data = pd.read_csv("E:\Term 6\CIS\Project 1\phase 2\Heart.csv")
heart_data.describe()
```

```python
heart_data = heart_data.dropna(axis = 0)
```

```python
heart_data.columns
```

```python
X = heart_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
X.head(15)
```

```python
y = heart_data.target
y.head(15)
```

```python
from sklearn.tree import DecisionTreeRegressor
heart_model = DecisionTreeRegressor(random_state = 1)
heart_model.fit(X, y)
```

```python
print(heart_model.predict(X.head(15)))
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(train_y, heart_model.predict(train_X)))
print("Test set Accuracy: ", metrics.accuracy_score(test_y, heart_model.predict(test_X)))
```
